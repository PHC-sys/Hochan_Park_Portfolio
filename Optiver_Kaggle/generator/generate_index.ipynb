```python
import pandas as pd
import numpy as np
import pickle
```

```python
#load
with open('../train.pickle', 'rb') as f:
    train_data = pickle.load(f)

#load
with open('../index.pickle', 'rb') as f:
    index_list = pickle.load(f)
    
#load
with open('../dataset_for_ML.pickle', 'rb') as f:
    dataset = pickle.load(f)
```

```python
train_data
```

```python
def wap_matched_size(train_data):
    index_list = []
    for i in range(0,481):
        print('date:',i)
        for j in range(0, 541, 10):
            data = train_data.query(f'date_id == {i} and seconds_in_bucket == {j}')
            index = (data['wap']*data['matched_size']).sum()/(data['matched_size']).sum()
            index_list.append(index)
    
    return index_list
```

```python
index_list_wap_matched = wap_matched_size(train_data)
```

```python
len(index_list_wap_matched), len(index_list)
```

```python
counter = 0
index_dic = {}
for date_id in range(481):
    for seconds_in_bucket in range(0, 541, 10):
        index_dic[(date_id, seconds_in_bucket)] = index_list[counter]
        counter += 1

print(counter)
print(index_dic)
```

```python
raw_data = train_data[['stock_id','date_id','seconds_in_bucket','wap','target']]
```

```python
raw_data = raw_data.dropna()
raw_data
```

```python
counter = 0
dataset = {}
for i in range(50):
    try:
        stock_data = raw_data.query(f'stock_id == {i}')
        stock_data['wap_60'] = stock_data['wap'].shift(-6)
        stock_data['index'] = np.NaN
        for date in stock_data.date_id.value_counts().sort_index().index:
            for second in stock_data.seconds_in_bucket.value_counts().sort_index().index:
                condition = (stock_data['date_id'] == date) & (stock_data['seconds_in_bucket'] == second)
                stock_data.loc[condition, 'index'] = index_dic[(date, second)]

        stock_data['index_60'] = stock_data['index'].shift(-6)
        stock_data = stock_data.dropna()
        print(f'stock_{i}: done')
    except:
        print(f'stock_{i}:exception')
    if counter == 0:
        print(stock_data)
    counter += 1
    dataset[i] = stock_data

# save
with open('dataset_for_ML_new_1.pickle', 'wb') as f:
    pickle.dump(dataset, f, pickle.HIGHEST_PROTOCOL)
```
